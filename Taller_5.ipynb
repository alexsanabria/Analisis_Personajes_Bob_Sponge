```python
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')

from nltk.stem.snowball import SnowballStemmer
spanishStemmer=SnowballStemmer("spanish")

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_distances
```

# Punto 1: Pre-Procesamiento

```python
data_bob=pd.DataFrame(pd.read_csv('bob_esponja.csv'))
data_bob
```

# Punto 2: TF-IDF

```python
data_bob['pre-procesado']=""
```

```python
cadena=""
for i in range(0, len(data_bob['Descripción'])):
    dep=data_bob['Descripción'][i] # Guardar el contenido de la celda en la variable dep
    dep=dep.lower() # Comverti a minuscula
    #dep=re.sub(r"[^a-zA-Z\s\d]", " ", dep)  
    dep=re.sub(r"[^a-zA-Záéíóúñ\s]", " ", dep) # Dejar solo letras
    dep=dep.split() # crear array con las palabras
    texto=[palabra for palabra in dep if palabra not in stopwords_sp] # Eliminar palabras vacias 
    Cadena=""
    for ele in texto:
        ele=spanishStemmer.stem(ele)
        espacio=" "
        cadena += espacio
        cadena += ele       
    data_bob['pre-procesado'][i]=cadena
    cadena=""
```

```python
data_bob
```

# Punto 3: Distancia del coseno

```python
index=list(data_bob['Personaje'])

tfidf_vect = TfidfVectorizer()
tfidf = tfidf_vect.fit_transform(data_bob['pre-procesado'].values)
tfidf_matrix = pd.DataFrame(tfidf.toarray(), columns=tfidf_vect.get_feature_names())
tfidf_matrix.index = index

tfidf_matrix.T.round(3)
```

```python
dist_cos = cosine_distances(tfidf_matrix.values)
dist_cos = pd.DataFrame(dist_cos, columns = tfidf_matrix.index, index = tfidf_matrix.index)
dist_cos
```

```python
dist_cos=dist_cos.round(3)
```

# ¿Cuáles son los personajes menos parecidos?
- Los personajes menos parecidos son la __Sra. Puff y Sheldon J. Plankton__ con un valor de 0.987 	

```python
Menos_parecidos = dist_cos[dist_cos == np.amax(dist_cos.values)] # Obtener el mayor valor en la matriz
Menos_parecidos
```

# ¿Cuáles son los personajes más parecidos?
- Los personajes más parecidos son __Eugenio H. Cangrejo y Sheldon J. Plankton__ con un valor de 0.688

```python
dist_cos.values[dist_cos.values==0]=2 #Reemplazar lo valores de 0 por un 2
```

```python
Mas_parecidos = dist_cos[dist_cos == np.amin(dist_cos.values)]# Buscar el valor menor en la matriz
Mas_parecidos
```
